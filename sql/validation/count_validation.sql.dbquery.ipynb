```sql
-- ============================================
-- Record Count Validation
-- Purpose: Verify row counts across Gold Layer
-- ============================================

SELECT 'dim_customers' AS table_name,
       COUNT(*) AS record_count
FROM fmcg.gold.dim_customers

UNION ALL

SELECT 'dim_products',
       COUNT(*)
FROM fmcg.gold.dim_products

UNION ALL

SELECT 'dim_date',
       COUNT(*)
FROM fmcg.gold.dim_date

UNION ALL

SELECT 'dim_gross_price',
       COUNT(*)
FROM fmcg.gold.dim_gross_price

UNION ALL

SELECT 'fact_orders',
       COUNT(*)
FROM fmcg.gold.fact_orders;
```
