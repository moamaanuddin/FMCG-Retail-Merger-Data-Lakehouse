```python
start_date='2024-01-01'
end_date='2025-12-01'
```

```python
df=spark.sql(f"""select explode (sequence(to_date('{start_date}'),to_date('{end_date}'),interval 1 month)) as month_start_date""")
display(df)

             
```

```python
from pyspark.sql import functions as F
upd_df=df.withColumn("date_key",F.date_format("month_start_date","yyyyMM").cast("int"))\
.withColumn("year",F.date_format("month_start_date","yyyy").cast("int"))\
.withColumn("month__name",F.date_format("month_start_date","MMMM"))\
.withColumn("month__short",F.date_format("month_start_date","MMM"))\
.withColumn("year_quater",F.concat(F.col("year"), F.lit("-Q"), F.quarter("month_start_date")))
display(upd_df)
```

```python
upd_df.write.mode("overwrite").format("delta").saveAsTable("fmcg.gold.dim_date")
```
