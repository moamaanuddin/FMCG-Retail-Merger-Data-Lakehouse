#Creating and setting up catalogs

```python
%sql
create catalog if not exists fmcg;
use catalog fmcg;
```

```python
%sql
create schema if not exists fmcg.bronze;
create schema if not exists fmcg.silver;
create schema if not exists fmcg.gold;

```

```python
%sql
use fmcg.gold;
alter table fmcg.gold.dim_customers alter column customer_code type string;
```

```python
%sql
select count(1) from fmcg.gold.fact_orders;
```
